```python
import matplotlib.pyplot as plt
import numpy as np

data = {
    "bin": [
        "Mistral-7B",
        "Shisa-gamma-7b",
        "Arithmo2-Mistral-7B",
        "Abel-7B",
        "MERGE3",
        "MERGE3",
        "MERGE3",
        "MERGE3",
        "EvoLLM-JP-7B"
    ],
    "accuracy": [0.0, 0.02, 0.036, 0.244, 0.344, 0.38, 0.38, 0.42, 0.488],
    "dataset_size": [5, 5, 5, 5, 20, 30, 50, 100, 1069],
    "flops": [1e-3, 1e-3, 1e-3, 1e-3, 15645, 23467, 39112, 78225, 4470000]
}

# Unique models (bins)
unique_bins = list(set(data["bin"]))
colors = plt.cm.tab10(np.linspace(0, 1, len(unique_bins)))

# Map model bins to colors
bin_to_color = {model: color for model, color in zip(unique_bins, colors)}

# Create the scatter plot
plt.figure(figsize=(10, 6))
for bin_name, accuracy, dataset_size, flops in zip(data["bin"], data["accuracy"], data["dataset_size"], data["flops"]):
    plt.scatter(
        flops, 
        accuracy, 
        s=dataset_size,  # Scale the size for visibility
        color=bin_to_color[bin_name], 
        alpha=0.5,  # Set transparency
        label=bin_name if bin_name not in plt.gca().get_legend_handles_labels()[1] else ""
    )

# Add labels and legend
plt.xlabel("FLOPs")
plt.ylabel("Accuracy")
plt.title("Model Accuracy vs FLOPs (Point Size = Dataset Size)")
#plt.xscale("log")  # Use log scale for FLOPs
plt.legend(title="Models", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

# Show plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = {
    "bin": [
        "Mistral-7B",
        "Shisa-gamma-7b",
        "Arithmo2-Mistral-7B",
        "Abel-7B",
        "MERGE3-20",
        "",
        "MERGE3-50\nMERGE3-30",
        "MERGE3-100",
        "EvoLLM-JP-7B"
    ],
    "accuracy": [0.0, 0.02, 0.036, 0.244, 0.344, 0.38, 0.38, 0.42, 0.488],
    "dataset_size": [5, 5, 5, 5, 20, 30, 50, 100, 1069],
    "flops": [1e-3, 1e-3, 1e-3, 1e-3, 15645, 23467, 39112, 78225, 4470000]
}

# Unique models (bins)
unique_bins = list(set(data["bin"]))
colors = plt.cm.tab10(np.linspace(0, 1, len(unique_bins)))

# Map model bins to colors
bin_to_color = {model: color for model, color in zip(unique_bins, colors)}

# Create the scatter plot
plt.figure(figsize=(10, 6))
for bin_name, accuracy, dataset_size, flops in zip(data["bin"], data["accuracy"], data["dataset_size"], data["flops"]):
    plt.scatter(
        flops, 
        accuracy, 
        s=dataset_size,  # Scale the size for visibility
        color=bin_to_color[bin_name], 
        alpha=0.5  # Set transparency
    )
    if bin_name == "EvoLLM-JP-7B":
        plt.text(
        flops - 0.01 * max(data["flops"]),  # Offset slightly to the right
        accuracy - 0.08 * max(data["accuracy"]), 
        bin_name, 
        fontsize=9, 
        ha="center", 
        va="center"
    )
    else:
        plt.text(
            flops + 0.02 * max(data["flops"]),  # Offset slightly to the right
            accuracy, 
            bin_name, 
            fontsize=9, 
            ha="left", 
            va="center"
        )

# Add labels
plt.xlabel("FLOPs")
plt.ylabel("Accuracy JPN-GSM8K")
#plt.title("Model Accuracy vs FLOPs (Point Size = Dataset Size)")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

# Show plot
plt.tight_layout()
plt.show()
```
